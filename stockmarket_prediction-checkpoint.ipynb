# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

# Loading the Dataset

```python
file_path = 'sp500.csv'
data = pd.read_csv(file_path)
```

```python
print("SF 500 Dataset:")
print(data.head())
```

# Preprossesing the Data 

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
features = ['Open', 'High', 'Low', 'Volume']
target = 'Close'
```

```python
data.dropna(inplace=True)
```

Feature Scaling

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data[features])
```

Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    scaled_features, data[target], test_size=0.2, random_state=42
)
```

# Linear Regression

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

# SVR

```python
svr_model = SVR(kernel='rbf')
svr_model.fit(X_train, y_train)
```

# Predictions and Evaluations

```python
lr_predictions = lr_model.predict(X_test)
svr_predictions = svr_model.predict(X_test)
```

```python
print("\nLinear Regression Performance:")
print("Mean Squared Error:", mean_squared_error(y_test, lr_predictions))
print("R-squared:", r2_score(y_test, lr_predictions))
```

```python
print("\nSupport Vector Regression (SVR) Performance:")
print("Mean Squared Error:", mean_squared_error(y_test, svr_predictions))
print("R-squared:", r2_score(y_test, svr_predictions))
```

```python
plt.figure(figsize=(14, 7))
```

```python
plt.subplot(1, 2, 1)
plt.plot(y_test.values, label='Actual', color='blue')
plt.plot(lr_predictions, label='Predicted', color='orange')
plt.title('Linear Regression - Actual vs Predicted')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(y_test.values, label='Actual', color='blue')
plt.plot(svr_predictions, label='Predicted', color='orange')
plt.title('SVR - Actual vs Predicted')
plt.legend()

plt.tight_layout()
plt.show()
```
