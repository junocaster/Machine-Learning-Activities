# Training and Evaluating the Models (Linear Regression and Random Forst Regression)

```python
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regression": RandomForestRegressor(n_estimators=100, random_state=42)
}

# Train and evaluate each model
results = {}
for model_name, model in models.items():
    try:
        # Train the model
        model.fit(X_train_scaled, y_train)
        
        # Predict on the test set
        y_pred = model.predict(X_test_scaled)
        
        # Evaluate model performance
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        results[model_name] = {'MSE': mse, 'R² Score': r2}
    except Exception as e:
        print(f"Error with {model_name}: {e}")
```

# Results and Visualizations of Each Models

```python
print("Model Performance Comparison:")
for model_name, metrics in results.items():
    print(f"{model_name} - MSE: {metrics['MSE']:.2f}, R² Score: {metrics['R² Score']:.2f}")
```
