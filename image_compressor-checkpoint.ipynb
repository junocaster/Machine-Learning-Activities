# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
from PIL import Image
```

# Load Image 

```python
image_path = 'nisah.jpg'
image = Image.open(image_path)
image = image.convert('RGB')  # Ensure the image is in RGB format
image_array = np.array(image)
```

# Reshape Image (for K-Means)

```python
rows, cols, ch = image_array.shape
image_flattened = image_array.reshape(-1, 3)
```

# K Means Clustering

```python
n_colors = 16 # number of clusters
```

```python
kmeans = KMeans(n_clusters=n_colors, random_state=42)
image_sample = shuffle(image_flattened, random_state=42)[:1000]  
kmeans.fit(image_sample)
```

Replace each pixel with its nearest cluster center

```python
cluster_centers = np.uint8(kmeans.cluster_centers_)
labels = kmeans.predict(image_flattened)
compressed_image = cluster_centers[labels]
compressed_image = compressed_image.reshape(rows, cols, 3)
```

```python
plt.figure(figsize=(12, 6))
```

Display Original Image

```python
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
```

Display Compressed Image

```python
plt.subplot(1, 2, 2)
plt.imshow(compressed_image)
plt.title('Compressed Image')
plt.axis('off')
```
