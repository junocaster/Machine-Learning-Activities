# Import Libraries

```python
import nltk
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from nltk.corpus import movie_reviews
from nltk.corpus import stopwords
import re
```

 # Download NLTK Data

```python
nltk.download('punkt')
nltk.download('movie_reviews')
nltk.download('stopwords')
```

# Preprocess Text Data

```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    # Use regular expressions to tokenize and clean the text
    tokens = re.findall(r'\b\w+\b', text.lower())  # Tokenize and lowercase
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    return ' '.join(tokens)
```

Get the movie reviews from nltk corpus

```python
documents = [(movie_reviews.raw(fileid), category) 
             for fileid in movie_reviews.fileids() 
             for category in movie_reviews.categories(fileid)]
```

Apply preprocessing

```python
documents = [(preprocess_text(text), label) for text, label in documents]
```

Split data into train and test datasets

```python
train_data, test_data = train_test_split(documents, test_size=0.2, random_state=42)
```

Separate text and labels

```python
train_texts, train_labels = zip(*train_data)
test_texts, test_labels = zip(*test_data)
```

Label negative review as 0 and positive review as 1

```python
train_labels = [1 if label == 'pos' else 0 for label in train_labels]
test_labels = [1 if label == 'pos' else 0 for label in test_labels]
```

Vectorize the text data using CountVectorizer

```python
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train_texts)
X_test = vectorizer.transform(test_texts)
```

# Naive Bayes Model

```python
nb_model = MultinomialNB()
nb_model.fit(X_train, train_labels)
```

Predictions and evaluation for Naive Bayes

```python
y_pred_nb = nb_model.predict(X_test)
print("Naive Bayes Model Evaluation:")
print(f"Accuracy: {accuracy_score(test_labels, y_pred_nb):.4f}")
print(f"Precision: {precision_score(test_labels, y_pred_nb):.4f}")
print(f"Recall: {recall_score(test_labels, y_pred_nb):.4f}")
print(f"F1 Score: {f1_score(test_labels, y_pred_nb):.4f}")
```

# Logistic Regression Model

```python
logreg_model = LogisticRegression(max_iter=200)
logreg_model.fit(X_train, train_labels)
```

Predictions and evaluation for Logistic Regression

```python
y_pred_lr = logreg_model.predict(X_test)
print("\nLogistic Regression Model Evaluation:")
print(f"Accuracy: {accuracy_score(test_labels, y_pred_lr):.4f}")
print(f"Precision: {precision_score(test_labels, y_pred_lr):.4f}")
print(f"Recall: {recall_score(test_labels, y_pred_lr):.4f}")
print(f"F1 Score: {f1_score(test_labels, y_pred_lr):.4f}")
```

# Plot Confusion Matrices

```python
def plot_confusion_matrices(y_true_nb, y_pred_nb, y_true_lr, y_pred_lr):
    cm_nb = confusion_matrix(y_true_nb, y_pred_nb)
    cm_lr = confusion_matrix(y_true_lr, y_pred_lr)

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # Create side-by-side plots
```

```python
def plot_confusion_matrices(y_true_nb, y_pred_nb, y_true_lr, y_pred_lr):
    cm_nb = confusion_matrix(y_true_nb, y_pred_nb)
    cm_lr = confusion_matrix(y_true_lr, y_pred_lr)

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # Create side-by-side plots

    # Naive Bayes Confusion Matrix
    sns.heatmap(cm_nb, annot=True, fmt='d', cmap='Reds', ax=axes[0], xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
    axes[0].set_title("Naive Bayes")
    axes[0].set_xlabel("Predicted")
    axes[0].set_ylabel("Actual")

    # Logistic Regression Confusion Matrix
    sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Reds', ax=axes[1], xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
    axes[1].set_title("Logistic Regression")
    axes[1].set_xlabel("Predicted")
    axes[1].set_ylabel("Actual")

    plt.tight_layout()
    plt.show()
```

```python
plot_confusion_matrices(test_labels, y_pred_nb, test_labels, y_pred_lr)
```
