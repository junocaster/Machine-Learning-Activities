# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

# Loading the dataset

```python
df = pd.read_csv('diabetes.csv')
```

```python
print(df.head())
```

```python
imputer = SimpleImputer(strategy='median')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

# Defining and Standardizing Features for scaling

```python
X = df_imputed.drop(columns=['Outcome', 'Glucose'])  # Features (exclude Outcome and Glucose)
y_class = df_imputed['Outcome']  # Target for classification (Diabetes prediction)
y_reg = df_imputed['Glucose']  # Target for regression (Glucose level prediction)
```

```python
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_class_scaled = scaler.fit_transform(X_train_class)
X_test_class_scaled = scaler.transform(X_test_class)
X_train_reg_scaled = scaler.fit_transform(X_train_reg)
X_test_reg_scaled = scaler.transform(X_test_reg)
```

# Training and Evaluating Classification Models

```python
classification_models = {
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier()
}
```

```python
classification_results = {}
for model_name, model in classification_models.items():
    try:
        model.fit(X_train_class_scaled, y_train_class)
        y_pred_class = model.predict(X_test_class_scaled)
        accuracy = accuracy_score(y_test_class, y_pred_class)
        classification_results[model_name] = accuracy
    except Exception as e:
        print(f"Error with {model_name}: {e}")

# Display classification results
print("\nClassification Model Accuracy:")
for model_name, accuracy in classification_results.items():
    print(f"{model_name}: {accuracy:.2f}")

# Regression models
regression_models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regressor": RandomForestRegressor(n_estimators=100, random_state=42)
}
```

```python
y_pred_class_lr = classification_models["Logistic Regression"].predict(X_test_class_scaled)
y_pred_class_knn = classification_models["KNN"].predict(X_test_class_scaled)
y_pred_class_dt = classification_models["Decision Tree"].predict(X_test_class_scaled)
```

# Classification Results

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
plt.scatter(y_test_class, y_pred_class_lr, color='blue', edgecolors='k', alpha=0.7)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.title('Actual vs Predicted (Logistic Regression)')
plt.xlabel('Actual Outcome')
plt.ylabel('Predicted Outcome')

plt.subplot(1, 3, 2)
plt.scatter(y_test_class, y_pred_class_knn, color='green', edgecolors='k', alpha=0.7)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.title('Actual vs Predicted (KNN)')
plt.xlabel('Actual Outcome')
plt.ylabel('Predicted Outcome')

plt.subplot(1, 3, 3)
plt.scatter(y_test_class, y_pred_class_dt, color='purple', edgecolors='k', alpha=0.7)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.title('Actual vs Predicted (Decision Tree)')
plt.xlabel('Actual Outcome')
plt.ylabel('Predicted Outcome')

plt.tight_layout()
plt.show()
```

# Training and Evaluating Regression Models

```python
regression_results = {}
for model_name, model in regression_models.items():
    try:
        # Fit the model on the training data
        model.fit(X_train_reg_scaled, y_train_reg)
        
        # Make predictions
        y_pred_reg = model.predict(X_test_reg_scaled)
        
        # Evaluate performance using MSE and R score
        mse = mean_squared_error(y_test_reg, y_pred_reg)
        r2 = r2_score(y_test_reg, y_pred_reg)
        
        # Store the results
        regression_results[model_name] = {'MSE': mse, 'R Score': r2}
        
    except Exception as e:
        print(f"Error with {model_name}: {e}")
```

# Regression Results

```python
print("Regression Model Performance:")
for model_name, metrics in regression_results.items():
    print(f"{model_name} - MSE: {metrics['MSE']:.2f}, R Score: {metrics['R Score']:.2f}")
```

```python
y_pred_reg_lr = regression_models["Linear Regression"].predict(X_test_reg_scaled)
y_pred_reg_rf = regression_models["Random Forest Regressor"].predict(X_test_reg_scaled)
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(y_test_reg, y_pred_reg_lr, color='blue', edgecolors='k', alpha=0.7)
plt.plot([min(y_test_reg), max(y_test_reg)], [min(y_test_reg), max(y_test_reg)], color='red', linestyle='--')
plt.title('Actual vs Predicted (Linear Regression)')
plt.xlabel('Actual Glucose')
plt.ylabel('Predicted Glucose')

plt.subplot(1, 2, 2)
plt.scatter(y_test_reg, y_pred_reg_rf, color='green', edgecolors='k', alpha=0.7)
plt.plot([min(y_test_reg), max(y_test_reg)], [min(y_test_reg), max(y_test_reg)], color='red', linestyle='--')
plt.title('Actual vs Predicted (Random Forest)')
plt.xlabel('Actual Glucose')
plt.ylabel('Predicted Glucose')

plt.tight_layout()
plt.show()
```
